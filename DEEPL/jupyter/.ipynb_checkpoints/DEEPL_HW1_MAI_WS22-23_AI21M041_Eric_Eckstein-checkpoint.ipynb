# Find the minimum of $f(x) = x^Tx$

## Tasks:

```python
# Provided, nothing to do here!: plotting function to visualize f(x)
# After optimization you can pass your tracking arrays to the function
def plot3d_quadratic(track_x=None, track_fx=None):
    plt.close()
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    
    # scale function for plotting
    SCALE = 1
    
    # offset fo better visualization
    OFFSET = 1
    
    
    # plot opimized path
    if track_x is not None and track_fx is not None:
        track_x = np.concatenate(track_x, axis=-1)
        track_fx = np.array(track_fx)
        ax.plot3D(track_x[0,:], track_x[1,:], np.zeros_like(track_x[1,:]), '-o', c='green')
        ax.plot3D(track_x[0,:], track_x[1,:], track_fx*SCALE+OFFSET, '-o', c='green')
    
    # Make grid
    X = np.linspace(-1, 1, 20)
    Y = np.linspace(-1, 1, 20)
    X, Y = np.meshgrid(X, Y)
    
    
    Z = (X**2 + Y**2)* SCALE + OFFSET

    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, 
                           linewidth=0, antialiased=False, alpha=0.5)
    ax.set_zlim(0, 2)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter('{x:.02f}')
    ax.contour(X, Y, Z, 30, zdir='z', offset=0, cmap=cm.coolwarm)
    

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
# Nothing to do here! Just execute!
# plot function without tracking arrays
plot3d_quadratic()
```

```python
# TODO!
# Task [A1]: Implement f(x) = x*x as a pytorch function
# For visualization purposes and simplicity we restrict the function to R²
def quadratic(x: torch.Tensor) -> torch.Tensor:
    assert len(x.shape) == 2 and x.shape[0] == 2 , f'Input must be a matrix of dimensions 2xN'
    x1 = x[0,:]
    x2 = x[1,:]
    ret = x1**2 + x2**2
    return ret
```

```python
# TODO!
# Task [A2]: Implement the gradient of the quadratic function as a pytorch function
# For visualization purposes and simplicity we restrict the function to R²
def grad_quadratic(x: torch.Tensor) -> torch.Tensor:
    assert len(x.shape) == 2 and x.shape[0] == 2 , f'Input must be a matrix of dimensions 2xN'
    x1 = x[0,:]
    x2 = x[1,:]
    gx1 = 2 * x1
    gx2 = 2 * x2
    ret = torch.stack((gx1, gx2))
    return ret
```

```python
# TODO!
# Task [A3]: Implement gradient descent with learning rate
# we already fix the starting point!
x = torch.tensor((-1.,-.1)).view(-1,1)

# set the learning rate!
lr = 0.025

# we fix the number of iterations
n_iter = 100

# we initialize our tracking arrays. Why do we need the clone function here? What does .item() do?
track_x = [x.clone()]
fx = quadratic(x)
track_fx = [fx.item()]

# implement gradient descent here!
for i in range(n_iter):
    print(f'\r iteration [{i+1}/{n_iter}], f(x) = {fx}', end="") 
    fx = quadratic(x) # Execute function 
    #print(f'\r f(x) = {fx}')
    g = grad_quadratic(x) # calculate gradient
    #print(f'\r g = {g}')
    x = x - lr * g # update x
    #print(f'\r x = {x}')
    track_x.append(x.clone())
    track_fx.append(fx.item())
```

```python
# Nothing to do here, exept executing!
# Plot function values over iterations:
plt.close()
plt.plot(track_fx)
plt.xlabel('Iteration')
plt.ylabel('f(x)')
plt.show()
```

```python
# TODO! 
# Task [A4]: pass your tracking arrays to the plot3d_quadratic function and plot your optimization path! Can you find a good learning rate to get to the minimum? 
plot3d_quadratic(track_x, track_fx)
```

```python
# Provided: plotting function to visualize Himmleblau's function
# After optimization you can pass your tracking arrays to the function
def plot3d_himmelblau(track_x=None, track_fx=None):
    plt.close()
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    
    # scale function for plotting
    SCALE = 0.001
    
    # offset fo better visualization
    OFFSET = 1
    
    
    # plot opimized path
    if track_x is not None and track_fx is not None:
        track_x = np.concatenate(track_x, axis=-1)
        track_fx = np.array(track_fx)
        ax.plot3D(track_x[0,:], track_x[1,:], np.zeros_like(track_x[1,:]), '-o', c='green')
        ax.plot3D(track_x[0,:], track_x[1,:], track_fx*SCALE+OFFSET, '-o', c='green')
    
    # Make grid
    X = np.linspace(-5.5, 5.5, 20)
    Y = np.linspace(-5.5, 5.5, 20)
    X, Y = np.meshgrid(X, Y)
    
    
    Z = ((X**2 + Y - 11)**2 + (X + Y**2 - 7)**2) * SCALE + OFFSET

    # Plot the surface.
    surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, 
                           linewidth=0, antialiased=False, alpha=0.5)
    ax.set_zlim(0, 2)
    ax.zaxis.set_major_locator(LinearLocator(10))
    ax.zaxis.set_major_formatter('{x:.02f}')
    ax.contour(X, Y, Z, 30, zdir='z', offset=0, cmap=cm.coolwarm)
    

    # Add a color bar which maps values to colors.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    
```

```python
plot3d_himmelblau()
```

```python
# TODO!
# Task [A5]: Implement himmelblaus function as a pytorch function
# For visualization purposes and simplicity we restrict the function to R²
def himmelblau(x: torch.Tensor) -> torch.Tensor:
    assert len(x.shape) == 2 and x.shape[0] == 2 , f'Input must be a matrix of dimensions 2xN'
    X = x[0,:]
    Y = x[1,:]
    ret = ((X**2 + Y - 11)**2 + (X + Y**2 - 7)**2) # TODO: implement function here!
    return ret
```

```python
# TODO!
# Task [A6]: Implement the gradient of himmelblau's function as a pytorch function
# For visualization purposes and simplicity we restrict the function to R²
def grad_himmelblau(x: torch.Tensor) -> torch.Tensor:
    assert len(x.shape) == 2 and x.shape[0] == 2 , f'Input must be a matrix of dimensions 2xN'
    X = x[0,:]
    Y = x[1,:]
    #gx1 = 2 *(-7 + X + 0**2 + 2 *X* (-11 + X**2 + 0)) # calculate x1 component here
    #gx2 = 2 *(-7 + 0 + Y**2 + 2 *0* (-11 + 0**2 + Y)) # calculate x2 component here
    #ret = torch.stack((gx1, gx2))
    return 2 *(-7 + X + 0**2 + 2 *X* (-11 + X**2 + 0))
    
```

```python
# TODO!
# Task [A6]: Implement gradient descent with learning rate
# we already fix the starting point! Can you reach all minima with this initial point?
x = torch.tensor((0.,-5.)).view(-1,1)

lr = 0.0025
n_iter = 100
track_x = [x.clone()]
fx = himmelblau(x)
track_fx = [fx.item()]

for i in range(n_iter):
    # add all necessary code here!
    fx = himmelblau(x) # Execute function 
    g = grad_himmelblau(x) # calculate gradient
    x = x - lr * g # update x
    
    #print(f'\r iteration [{i+1}/{n_iter}], x= {x}, f(x) = {fx}, g = {g}') 
    track_x.append(x.clone())
    track_fx.append(fx.item())
    track_x.append(x.clone())
    track_fx.append(fx.item())
```

```python
# Nothing to do here, exept executing!
# Plot function values over iterations:
plt.close()
plt.plot(track_fx)
plt.xlabel('Iteration')
plt.ylabel('himmelblau(x)')
plt.show()
```

```python
# TODO! 
# Task [A8]: pass your tracking arrays to the plot3d_quadratic function!
plot3d_himmelblau(track_x,track_fx)
```

```python
# TODO!
# Task [A9] [A10]: Implement the optimization via torch optim
# Try out SGD and Adam and various learning rates. What do you observe?

# - sgd seems to find the minimum faster than adam (for lr = 0.0005)
# - sgd seem to find minimum in less much less iterations (~10 vs ~200 on lr 0.005)

x = torch.tensor([[0.],[-5.]], requires_grad=True)
print(x.shape)

lr = 0.005
opt = torch.optim.SGD([x], lr=lr)# select optimizer from torch.optim and pass the appropriate parameters
#opt = torch.optim.Adam([x], lr=lr)
n_iter = 2000
# why do we need detach() ?
track_x = [x.detach().clone()]
track_fx = [himmelblau(x).detach().item()]
for i in range(n_iter):
    # implement necessary code here
    opt.zero_grad()
    fx = himmelblau(x)
    fx.backward()
    opt.step()
    
    track_x.append(x.detach().clone())
    track_fx.append(fx.detach().item())
```

```python
# Nothing to do here, exept executing!
# Plot function values over iterations:
plt.close()
plt.plot(track_fx)
plt.xlabel('Iteration')
plt.ylabel('himmelblau(x)')
plt.show()
```
